```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
bbc_df = pd.read_csv("data/bbc_news.csv")
```

### Exploratory Data Analysis

```python
# shape of the data
bbc_df.shape
```

```python
bbc_df.head()
```

```python
bbc_df.describe()
```

```python
# categorize the news based on the guid 
bbc_category = bbc_df['guid'].str.extract(r'/news/([a-z|-]+)-\d*')
bbc_category.head()
```

```python
# plot the distribution of news categories in bbc news
fig = plt.figure(figsize=[10,20])
ax = sns.countplot(y=bbc_category[0], palette='rocket', order=bbc_category[0].value_counts().index)
plt.title('Distribution of Categories')
plt.ylabel('Categories')
plt.xlabel('Count of News Articles')
```

```python
# the length of the news title
title_len = bbc_df['title'].str.len()
title_len.describe()
```

```python
# the length of the news description
description_len = bbc_df['description'].str.len()
description_len.describe()
```

```python
# tf-idf on the title 
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(bbc_df['title'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
max_tfidf_indices = tfidf_df.idxmax(axis=1)
title_tfidf = pd.DataFrame({'title':bbc_df['title'], 'key_word':max_tfidf_indices})
title_tfidf.head()
```
